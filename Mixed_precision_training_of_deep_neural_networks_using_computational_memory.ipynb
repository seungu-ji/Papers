# **Mixed-precision training of deep neural networks using computational memory**

논문 링크: https://arxiv.org/abs/1712.01192

본 논문은 nvidia blog를 뒤져보다 알게 되어 읽게 되었다.
(NVIDIA Blog 정리글 링크: https://developer.nvidia.com/blog/mixed-precision-training-deep-neural-networks/)

Multilayer perceptrons나 CNN, LSTM 등을 비롯한 모든 deep neural network(DNN)들은 모두 뉴런이라고 불리는 병렬 처리 장치를 통해 상호 연결의 weight를 조정함으로써 기존 알고리즘으로 해결하기 힘든 문제들을 해결한다.

이런 DNN들은 모두 훈련 과정을 거치게 되는데, 훈련 단계에서 무차별 대입(brute force) 접근 방식은 GPU와 같은 컴퓨팅 리소스의 가용성에도 불구하고 시간이 오래 걸린다.

따라서 이러한 부분을 개선시키기 위해 최근에는 DNN을 위한 non-von Neumann co-processor 설계에 많은 관심과 연구가 이루어지고 있다. 관련 연구들에서 계속 주목하고 있는 분야는 in-memory computing혹은 computational memory이다. 이 중 본 논문에서는 메모리 단위의 제한된 정밀도에 집중을 하여 **mixed precision in-memory computing** 개념을 제안한다.
