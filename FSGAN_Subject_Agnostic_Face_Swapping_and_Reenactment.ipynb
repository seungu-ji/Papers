반면 **face reenactment란, 위의 그림 Figure 1의 우측 부분과 같이 source image의 얼굴이 target video의 얼굴 움직임과 똑같이 자연스럽게 움직이는 작업**이다.

이 두 작업은 모두 엔터테인먼트나 개인 정보 보호 등을 위해 여러 관심을 받고 있는 분야들이다.

## **FSGAN 구조**

기존 이 분야는 Face2Face나 3DMM과 같은 3D 관련 기술이나 주로 Pix2Pix나 CycleGAN과 같은 혹은 GANimation, GANnotation 등 GAN 기술로 커버를 하고 있었지만, 본 논문은 기존 기술보다 더 뛰어난 성능을 얻기 위해 FSGAN을 제안하게 되었다. 본 논문의 contribute(기여)는 다음과 같다.

- Subject agnostic swapping and reenactment: FSGAN은 높은 품질과 시각적으로 일관된 결과물을 제공하며 동시에 person-specific 혹은 pair-specific train없이 포즈, 표현 및 정체성을 모두 고려한다.
- Multiple view interpolation: **Delaunay Triangulation** 및 무게 중심 좌표 등을 기반으로 동일한 얼굴의 여러 뷰를 연속적으로 보간하는 새로운 방식을 제공한다.
- New loss functions: 본 논문에서는 총 두 가지의 새로운 loss function을 제안한다. 첫 번째는 작은 단계에서 점진적으로 얼굴 재연을 훈련하기 위한 stepwise consistency loss이며, 두 번째는 source 얼굴을 새로운 context에 매끄럽게 통합하도록 face blending network를 훈련시키기 위한 Poisson blending loss이다.
