# **An Image is Worth 16x16 Words: Transformers for Image Recognition at Scale**

논문 링크: https://arxiv.org/abs/2010.11929

Transformers는 NLP에서 선택되는 모델이 되었다. 대부분의 접근 방식은 large text corpus에서 pre-train을 수행하고 task-specific dataset에 대해 fine-tuning을 수행하는 것이다(BERT). Transformers의 계산 효율성 및 확장성으로 100B 이상의 parameter를 사용하여 전례없는 크기의 모델을 학습 할 수 있게 되었다.

하지만 VIsion 분야에서는 아직 CNN을 많이 사용한다. NLP 분야의 성공에 영감을 받은 여러 연구들에서는 CNN과 유사한 architecture를 self-attention과 결합하려고 시도하여 '**Stand-Alone Self-Attention in Vision Models**'이나 **Axial-DeepLab: Stand-Alone Axial-Attention for Panoptic Segmentation**' 등 일부는 CNN을 완전히 대체한다.

후자의 연구들은 이론저그로 꽤나 효율적이지만, **specialized attention pattern**을 사용하기 때문에 최신 GPU에서는 아직 효과적으로 사용하기 힘들다. 따라서 large-scale image recognition task에서는 아직도 ResNet 계열의 architecture가 state-of-the-art 성능을 보여준다.
