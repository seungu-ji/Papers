# **PointNet: Deep Learning on Point Sets for 3D Classification and Segmentation**

논문 링크: https://arxiv.org/abs/1612.00593

해당 논문은 3D point cloud 데이터르ㄹ NN의 학습 데이터로 사용하려고 했을 때 발생하였던 문제점들을 해결한 논문이다.

image.png

기본적으로 point cloud와 같은 mesh type의 데이터는 image와 달리 정형화되어 있지 않다. 즉, 모든 공간이 가득 차 있는 것이 아니라 위의 그림의 (b)와 같이 각 점들이 매우 sparse하게 분포되어 있다.

따라서 본 논문이 나오기 전까지 기존의 논문들은 빈 공간까지 의미론적으로 채우는 3D 복셀화를 하거나 2D로 projection하여 projection된 정보를 사용하였다. 하지만 이러한 translation 기법들은 데이터를 불필요하게 대량으로 불리는 것이기 때문에 좋은 방법이 아니다.

이러한 문제점들로 인해 PointNet은 데이터를 그냥 바로 direct하게 input할 수 있게 설계했다.
