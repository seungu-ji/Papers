# **Training BatchNorm and Only BatchNorm: On the Expressive Power of Random Features in CNNs**

논문 링크: https://arxiv.org/abs/2003.00152

StyleGAN에서 사용된 AdaIN이나 multitask learning 등에 이르기까지 다양한 NN 기술은 feature의 아핀 변환(affine transformation)에 의존을 많이 한다. 그 중 가장 눈에 띄는 것은 BN인데, 본 논문에서는 BN의 β와 γ에 포커스를 맞췄다.

image.png

BN에 대해 이미 알고 있는 한 가지 사실은 위의 그림 Figure 1에서 볼 수 있듯이 ResNet의 성능에 의미있는 영향을 미쳐 acc를 0.5% ~ 2% 가량 향상시키며, ImageNet에 대해서도 2%가량 acc를 향상시킨다는 것이다.

이처럼 BN을 자주 사용하지만, BN의 scaling & shifting parameter인 γ와 β의 정확한 역할은 아직 알려지지 않았다.

따라서 본 논문에서는 다양한 width와 depth의 ResNet에서 parameter를 random initialization한 후 freeze한다. 그리고 BN의 parameter인 γ와 β만 학습을 진행한다. 이렇게 학습을 진행할 시 network는 여전히 동일한 기능을 유지하지만, 매우 적은 parameter만 학습을 한다.(최대 0.6%)

본 논문은 애초에 SOTA 정확도를 목표로 한 것이 아니다. 본 논문에서는 연구를 affine BN parameters의 표현력과 중요성을 보여준다.
