# **FCOS: Fully Convolutional One-Stage Object Detection**

논문 링크: https://arxiv.org/abs/1904.01355

image.png

FCOS는 **anchor-box를 사용하지 않는 one-stage detector**이다. FCOS는 기본적으로 anchor-based detector의 단점을 해결하고자 나온 모델이며, 위의 그림 Figure 1과 같이 **bbox의 4면으로부터 center keypoint까지의 거리인 (l, t, r, b)를 예측**하고 FPN의 각 level마다 거리(l, t, r, b)를 제한하여 multi-level prediction을 한다. 또한, **낮은 quality의 bbox를 제거하기 위한 center-ness**를 제안한다.

먼저, 간단하게 anchor-box를 사용하게 되면 생기는 단점들에 대해 소개해보겠다.
1. detection 성능이 anchor-box의 hyper-parameter인 scale, aspect ratios 등에 민감하다.
2. 미리 정의된(pre-defined) anchor-box가 generalization 성능에 악역향을 끼친다. anchor box와 다른 size나 aspect ratios를 가지고 있는 object를 detect하기 힘들다.
3. high recall rate를 위해 많은 anchor-box를 사용하게 되는데, 당연히 이 중에 수 많은 anchor box들이 negative에 할당이 되므로 class imbalance 문제가 발생한다.
4. anchor box와 GT의 IoU를 계산하는 복잡한 계산 과정이 필요하다.
