구조를 어떻게 변형했는지는 위의 그림 Figure 1을 보면 직관적으로 알 수 있다. ResNeXt도 VGG나 ResNet처럼 같은 layer들을 반복하는 것인데, 추가적으로 위의 그림 Figure 1에서 오른쪽 부분과 같이 GoogLeNet의 Inception과 같이 **1개의 input을 여러 방향으로 보내는 split transform merge(분할, 변형, 병합)** 방식을 이용한다.

Inception-ResNet과의 차이점은 각 path별로 같은 layer 구성을 가지고 있다는 점이다. 이를 grouped convolution이라고 한다.

위의 표 Table 1은 ResNet-50과 ResNeXt-50의 구성이다. 둘의 비교를 위해 params과 FLOPs를 비슷하게 유지해놓은 것을 볼 수 있다. ResNet은 하나의 convolution을 깊은 채널로 만들며 반면에 ResNeXt는 조금 더 깊지만 32개의 group convolution을 통해 연산량을 크게 낮추었다. (위의 표에서 C는 cardinarity로, group convoltion의 갯수이다.)

ResNet의 경우는 블록을 하나 쌓을 때, convolution을 2개만 진행을 했다. 하지만 ResNeXt의 경우는 2개의 블록만 쌓게 된다면 group convolution의 의미가 없어져 성능 향상에 의미가 없게 된다. 따라서 ResNeXt에서는 block의 depth가 3 이상일 때 부터 성능이 향상된다.
