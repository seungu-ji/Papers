# **Dynamic Routing Between Capsules**

논문 링크: https://arxiv.org/abs/1710.09829

CapsNet 혹은 Capsule Network이라고 불리는 이 network란 무엇일까? 이 network가 Convolution Neural Network보다 성능이 더 우수하나? 먼저 이를 알아보기 전에 왜 이러한 아이디어가 나왔는지에 대해 알아볼 필요가 있다.

일단 기존 CNN은 근본적으로 데이터의 본질을 잃지 않게 convolution layer를 쌓아 많은 행렬 곱셈을 진행한다. 이를 좀더 세분화하여 설명하자면, input image가 주어지면 convolution 작업을 수행하여 feature map을 생성하고 activation 및 pooling 작업을 차례로 통과한 후 이러한 작업을 반복하여 최종적으로 softmax classifier와 같은 분류기를 통해 분류를 진행한다.

이때, **pooling의 개념은 각 하위 영역(sub-region)을 요약하고 위치 정보와 관련하여 invariance를 줄 수 있다**는 것이 핵심 개념이다.

pooling이라는 것을 대입하게 된 계기가 바로 CNN의 한계점 때문이다. CNN은 **Visualizing and Understanding Convolutional Networks** 논문에 나와있듯이 처음에는 간단한 특징을 포착하고 layer가 깊어질수록 점점 복잡한 특징들을 잡아낸다. 그러나 이때 **더 깊은 layer(higher layer)들은 단순히 얕은 layer(lower layer)들의 가중합(weighted sum)**이라는 것이 가장 큰 문제점이다. 왜냐하면 이러한 **weighted sum은 simple feature와 complex feature간의 위치 관계를 전혀 고려하지 않기 때문**이다.

따라서 이러한 CNN의 문제점을 해결하기 위해 max pooling이라는 개념을 도입한 것이다. max pooling으로 인하여 feature map들의 spatial size를 줄어주어 higher layer들의 시야각(field of view)를 넓혀준다. 시야각을 넓혀준다는 것은 간단히 생각해 fps 게임이나 기타 게임에서 시야를 넓어주는(눈 사이의 길이가 더 길어진다는 느낌?) 개념과 유사하다고 생각하면 된다.
